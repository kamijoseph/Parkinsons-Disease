# Parkinsons Disease Prediction With SVM

```python
# dependencies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
data = pd.read_csv("parkinsons/parkinsons.csv")
data.head()
```

```python
# checking for missing values
data.isnull().sum()
```

```python
# dropping the name column
data = data.drop("name", axis=1)
data.head()
```

```python
data.groupby("status").mean()
```

```python
# feature and targets separation
X, y = data.drop("status", axis=1), data["status"]
X
```

```python
# splitting into training and testing data
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=3
)
print(X.shape, X_train.shape, X_test.shape)
```

```python
# scaling the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
print(X_train_scaled)
```

```python
# model training
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
```

```python
# model evalaution

# accuracy on the training data
X_pred = model.predict(X_train_scaled)
X_acc = accuracy_score(y_train, X_pred)
print(X_acc)
```

```python
# accuracy on the test data
y_pred = model.predict(X_test_scaled)
y_acc = accuracy_score(y_test, y_pred)
print(y_acc)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred = model.predict(X_test)

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
